### Numpy - Sort  [`numpy.sort()`](https://numpy.org/doc/stable/reference/generated/numpy.sort.html)   

Return a sorted copy of an array.

```python
import numpy as np
```

```python
N = np.random.randint(1,100,10)

print(N)
```

```python
O = np.random.randint(1,100,24).reshape(6,4)
print(O)
```

```python
np.sort(N)
```

```python
np.sort(O)
```

```python
np.sort(O,axis=1)
```

```python
np.sort(O,axis=0)
```

### NUMPY - APPEND  [`numpy.append()`](https://numpy.org/doc/stable/reference/generated/numpy.append.html)  

The numpy.append() appends values along the mentioned axis at the end of the array

```python
np.append(N,100)
```

```python
O
```

```python
np.append(O,np.array([[100,101,102,103]]),axis=0)  # axis = `0` means rows in numpy

np.append(O,np.ones((O.shape[0],1)),axis=1)      # axis = `1` means columns in numpy
```

### NUMPY - CONCATENATE  [`numpy.concatenate()`](https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html)

numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

```python
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print(b)
```

```python
np.concatenate((a,b),axis=0)  # axis = `0` means rows in numpy
```

```python
np.concatenate((a,b),axis=1)  # axis = `1` means columns in numpy
```
